```python
import pandas as pd
import numpy as np
```

```python
# Membaca dataset
df = pd.read_csv('covid_19_indonesia_time_series_all.csv')
df.head()
```

```python
# Melihat informasi tipe data
df.info()
```

```python
# Melihat nama kolom
df.columns
```

```python
# Menampilkan kolom missing values
nan_df = df[df.isna().any(axis=1)]
display(nan_df.head())
```

```python
# Cek nilai berdasarkan tipe data
df.select_dtypes('object')
```

```python
# Cek nilai berdasarkan tipe data
df.select_dtypes('int64')
```

```python
# Cek nilai berdasarkan tipe data
df.select_dtypes('float64')
```

```python
df.drop(["Location ISO Code", "New Cases", "New Deaths", "New Recovered", "New Active Cases", "Province", "Country", "Continent", "Island", "Time Zone", 
                "Special Status", "Total Regencies", "Total Cities", "Total Districts", "Total Urban Villages", "Total Rural Villages", "Area (km2)", 
                "New Cases per Million", "Total Cases per Million","New Deaths per Million", "Total Deaths per Million", "Case Fatality Rate",
                "Case Recovered Rate", "Growth Factor of New Cases","Growth Factor of New Deaths", "City or Regency"],axis=1,inplace=True)
```

```python
df.drop(df[(df['Location']=='Indonesia')].index,inplace=True)
```

```python
# Membuat index baru
df = df.reset_index()
```

```python
# Menghapus index lama
df = df.drop(columns='index')
```

```python
# Menyimpan file terbaru
df.to_excel('dataclean_c-19.xlsx')
```
